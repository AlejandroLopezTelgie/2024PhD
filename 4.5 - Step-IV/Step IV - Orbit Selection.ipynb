## 5.4 Step IV.- Orbit Selection

At this this stage, given the area of interest, both orbits do provide coverage (the final one is to be assessed based on orbital analysis and mission lifetime). SSO being higher do present an advantage in orbital lifetime and coverage.
Regarding launch opportunities, literature and launch providers suggest a minimum lead time in the order of 6 months for a brokered POD launch while SpaceX has a lead time in the order of 24 months. This is particularly relevant to 2nd generation, as the mission is required to be operational in 3 years (Req 3.14 schedule).
As for price, Spaceflight is the only broker which had available prices, see Table 2 7 for further details. In summary the order is 100 kUSD for 1U and 150 kUSD for a 3U. 
No orbit manoeuvring (raising) capabilities are considered at this stage, as the current technology readiness level of propulsion for CubeSat is not regarded as mature enough nor able to raise significantly with respect to the orbit where it is deployed by the launch vehicle (Yet several companies, like D-Orbit, and subsystems are currently developing these capabilities. Hence this must be re visit at the time of running the algorithm/heuristic).

With the goal of providing all the orbital inputs for the scenario simulation, the orbit inputs are:

- Orbits: Sun synchronous (SSO) and/or International Space Station (ISS) 
- Altitudes
    - Maximum altitude for SSO orbit is constrained to 600 km to allow for passive deorbit within 25 years $h_{SSO,max}=600 [km]$ 
	- Minimum altitude from cost-effective orbits is set as $h_{SSO,min}=450 [km]$ 
- Inclination: for SSO is a function of altitude given by Eq. 5 1.

$$ 
    \cos i \approx -\frac{2\rho}{3 J_2 R_\text{E}^2 \sqrt{\mu_E}} a^{7/2} [-] 
$$  
Eq. 5-1 (OR Eq. C-34 New SMAD)

$$
i_{SSO}=\cos^{-1} { (-4.77351328\times 10^{-15}a^{7/2}) }
$$


Where the variables are defined in Table 5-3, and yields inclination of SSO as

- 97.87[°] for $h_{SSO,max}$ 
- 97.21[°] for $h_{SSO,min}$

For ISS at an inclination of 51.64 [°] we have three altitudes at which to assess the orbital scenario:
- $h_{ISS,max}=460 [km]$ Maximum altitude for ISS-like satellites (likely deployed by the launch vehicle at this higher altitude as the ones deploy from ISS are deployed towards earth i.e. lower).
- $h_{ISS,min}= 275 [km]$ Note that at this altitude re-entry will happen in a matter of months.
- $h_{ISS,sweetspot}=350 [km]$ Note that ISS will deploy towards Earth (i.e. below the station to avoid debris collision risk), thus this is feasible only if deployed once the station is above.

### Data rate estimation for payload
These parameters are also used to size the payload generation data rate  $\dot{D}_{gen}$ from Eq. 5 2: 

$$ {\dot{D}}_{gen}=\frac{S_wV_g}{R_{GSD}b}\left[\frac{bit}{s}\right]  $$  
Eq. 5-2 



Where:
- $S_w$ is the swath width [m] set as 25 [km] by analogy from KITSUNE forest monitoring CubeSat mission (AZAMI et al., 2022a) 
- $b$ is the bit rate [bit/s]. Set as 12 [bit/s] by analogy from KITSUNE forest monitoring CubeSat mission (AZAMI et al., 2022a)
- $R_{GSD}$ is the ground sample distance (GSD) resolution [m], estimated previously based on the relationships by (VALENZUELA; REYES, 2019c). Note that the case is set to start with a $R_GSD$ of 25 m (while the required resolution is 50 m) to leave margin. As stated before, from the orbits in question GSD of 5 m are feasible.


#### Review of ground speed

- Formerly

- $V_g$ is ground speed [m/s], estimated as follows [New SMAD - Appendix C] in Eq. 5-6:

$$ V_g = \sqrt{\frac{\mu_E}{R_E + h}} \left [\frac{m}{s} \right ] $$
Eq. 5-3

- Now

$$ V_g=\frac {2\pi R_E}{P} $$

Where $P$, the orbital period is given by Kepler's third law as

$$ P= 2 \pi \sqrt{\frac{a^3}{\mu_E}}$$

Which yields

$$ V_g=R_E \sqrt{\frac{\mu_E}{a^3}}$$

---- 



Where the constants $\mu_E$ and $R_E$ have been defined in Table 5-3, and h represents the different altitudes. For and SSO at 600 km altitude, with a GSD of 25 m and a swath width of 25 km, and 12 bit, it yields 91.69 [Mbit/s].

To quantify the amount of generated data, the previously calculated data rate ${\dot{D}}_{gen}$ and the median contact time over the area of interest $t_{AoI}$ are used such that: 

$$ D_{gen}={\dot{D}}_{gen}t_{AoI}[Mbit] $$
Eq. 5-4

As well to quantify the amount of downloaded data, the download data rate $\dot{D_{DL}}$ and the median contact time to the ground station $t_{GS}$ are used as shown in Eq. 5-5, where a factor of 0.8 was implemented to account for communication protocols: 

$$D_{DL}=0.8{\dot{D}}_{DL}t_{GS}[Mbit]$$ 
Eq. 5-5

Both times $t_{AoI}$ and $t_{GS}$ are estimated from the orbital simulation implemented in FreeFlyer. The ground station time is further constrained by taking only the passes above 5 mins.
